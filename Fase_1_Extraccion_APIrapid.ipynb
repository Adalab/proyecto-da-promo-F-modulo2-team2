```python
print(lista_peliculas[0]["results"])
```

```python
lista_movie_segunda_tanda_comedy= []
for peli in lista_peliculas:
    for i in peli["results"]:
        lista_todas_pelis = []
        tipo_peli = i['titleType']["text"]
        nombre_peli = i['titleText']["text"]
        id_peli =i['id']
        lista_todas_pelis.append(tipo_peli)
        lista_todas_pelis.append(nombre_peli)
        lista_todas_pelis.append(id_peli)
        
        try: 
            mes_estreno_peli = i["releaseDate"]["month"]
        except:
            mes_estreno_peli = ''
        lista_todas_pelis.append(mes_estreno_peli)
        try:
            ano_estreno_peli = i["releaseDate"]["year"]
        except:
            ano_estreno_peli = ''
        lista_todas_pelis.append(ano_estreno_peli)
        lista_movie_segunda_tanda_comedy.append(tuple(lista_todas_pelis))
    
print(lista_movie_segunda_tanda_comedy)
```

```python
df_lista_movie_segunda_tanda_comedy = pd.DataFrame(lista_movie_segunda_tanda_comedy,columns = ["Tipo", "Título", "ID", "Mes Estreno",  "Año Estreno"])
print(df_lista_movie_segunda_tanda_comedy)
```
